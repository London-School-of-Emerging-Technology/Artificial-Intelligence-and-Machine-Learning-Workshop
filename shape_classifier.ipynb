```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Define shape data (features: [sides, angles, curves])
X = np.array([
    [0, 0, 1],   # Circle
    [3, 60, 0],  # Triangle
    [4, 90, 0],  # Square
    [0, 0, 1],   # Circle
    [4, 90, 0],  # Square
    [3, 60, 0]   # Triangle
])

# Step 2: Define labels (0: Circle, 1: Triangle, 2: Square)
y = np.array([0, 1, 2, 0, 2, 1])

# Step 3: Split data into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train the model using K-Nearest Neighbors (KNN)
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# Step 5: Make a prediction for a new shape
new_shape = np.array([[3, 60, 0]])  # Example: 3 sides, 60° angles, no curves (Triangle)
prediction = model.predict(new_shape)

# Step 6: Print the prediction result
shapes = {0: "Circle", 1: "Triangle", 2: "Square"}
print(f"The AI predicts: {shapes[prediction[0]]}")

# Step 7: Visualizing the shape classification
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=100)
plt.xlabel("Sides")
plt.ylabel("Angles")
plt.title("Shape Classification (AI Prediction)")
plt.colorbar(label="Shape Type (0: Circle, 1: Triangle, 2: Square)")
plt.show()
```
