```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random

# üìå Generate training data (e.g., savings over weeks)
weeks = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
savings = np.array([5, 10, 15, 20, 25, 30])  # Savings increase by ¬£5 per week

# Train a Linear Regression model
model = LinearRegression()
model.fit(weeks, savings)

# üéØ Game Start
print("\nüéØ Welcome to the Regression Prediction Game!")
print("Can you predict how much money will be saved in a future week?\n")

# üìå Show the past data as a hint
print("üí° Hint: Here are past savings records:")
for i in range(len(weeks)):
    print(f"üìÖ Week {weeks[i][0]}: ¬£{savings[i]}")

# Choose a random future week
future_week = random.randint(7, 12)
print(f"\nüìä The pattern suggests savings are increasing by a fixed amount each week.")
print(f"üí≠ Think carefully and predict the savings for **Week {future_week}**!")

# Ask the player for a guess
player_guess = float(input(f"\nüí° Your Guess for Week {future_week}: ¬£"))

# Model Prediction
predicted_savings = model.predict(np.array([[future_week]]))[0]

# üéâ Show Results
print(f"\nüì¢ Actual Regression Prediction: ¬£{round(predicted_savings, 2)}")
print(f"üìä Your Guess: ¬£{player_guess}")

# Scoring and Feedback
error = abs(predicted_savings - player_guess)

if error == 0:
    print("üéâ Perfect! You predicted it exactly right! ‚úÖ")
elif error < 5:
    print("üëç Great job! Your prediction was very close! üî•")
elif error < 10:
    print("üôÇ Good attempt! You are getting better at spotting patterns! üîÑ")
else:
    print("‚ùå Oops! Try again. Remember to look at the past data carefully! üöÄ")

# üìà Show Graph
plt.scatter(weeks, savings, color='blue', label="Past Data")
plt.plot(weeks, model.predict(weeks), color='green', label="Regression Line")
plt.scatter([future_week], [predicted_savings], color='red', label="Prediction")
plt.xlabel("Weeks")
plt.ylabel("Savings (¬£)")
plt.title("Regression Prediction Game")
plt.legend()
plt.show()
```
